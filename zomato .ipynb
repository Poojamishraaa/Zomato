```python
 import pandas as pd 
# Merge the datasets based on the country code
df1 = pd.merge(df1, df2, on='Country Code', how='left')

# Check for missing values in the merged dataset
df1.isnull().sum()

# Drop unnecessary columns
df1= df1.drop(['City', 'Address', 'Locality Verbose', 'Switch to order menu'], axis=1)

# Check the updated dataset
df1.head()
```

```python
# Analyzing average cost for two
plt.figure(figsize=(12, 6))
sns.boxplot(x='Average Cost for two', y='Country Code', data=df1)
plt.title('Average Cost for Two in Different Countries')
plt.xlabel('Average Cost for Two')
plt.ylabel('Country')
plt.show()

# Analyzing price range
plt.figure(figsize=(10, 6))
sns.countplot(x='Price range', data=df1)
plt.title('Distribution of Price Range')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Prepare the data for modeling
X = df1.drop(['Average Cost for two', 'Price range'], axis=1)
y_cost = df1['Average Cost for two']
y_price_range = df1['Price range']

# Split the data into training and testing sets
X_train, X_test, y_cost_train, y_cost_test = train_test_split(X, y_cost, test_size=0.2, random_state=42)
X_train, X_test, y_price_range_train, y_price_range_test = train_test_split(X, y_price_range, test_size=0.2, random_state=42)

# Initialize and train the linear regression models
cost_model = LinearRegression()
price_range_model = LinearRegression()
cost_model.fit(X_train[['Votes', 'Aggregate rating']], y_cost_train)
price_range_model.fit(X_train[['Votes', 'Aggregate rating']], y_price_range_train)

# Predict the average cost for two and price range
y_cost_pred = cost_model.predict(X_test[['Votes', 'Aggregate rating']])
y_price_range_pred = price_range_model.predict(X_test[['Votes', 'Aggregate rating']])

# Evaluate the models using mean squared error
cost_mse = mean_squared_error(y_cost_test, y_cost_pred)
price_range_mse = mean_squared_error(y_price_range_test, y_price_range_pred)

print('Mean Squared Error (Average Cost for two):', cost_mse)
print('Mean Squared Error (Price range):', price_range_mse)
```
